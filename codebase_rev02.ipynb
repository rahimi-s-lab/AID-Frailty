```python
import os
import pandas as pd



############ Data Cleansing P01
#Concatenate Daily datasets of both sensors and save each in a new csv file

# Directory containing your CSV files
p01_directory1 = './dataset/P01/Presence-80029C91F80E'
p01_directory2 = './dataset/P01/Presence-80029C91F820'

# Get a list of all CSV files in the directory
p01_csv_files_1 = [file for file in os.listdir(p01_directory1) if file.endswith('.csv')]
p01_csv_files_2 = [file for file in os.listdir(p01_directory2) if file.endswith('.csv')]

# Initialize an empty list to store dataframes
p01_dfs1 = []
p01_dfs2 = []

# Iterate through each CSV file, read it, and append it to the list
for file in p01_csv_files_1:
    df1 = pd.read_csv(os.path.join(p01_directory1, file))
    p01_dfs1.append(df1)
    
for file in p01_csv_files_2:
    df2 = pd.read_csv(os.path.join(p01_directory2, file))
    p01_dfs2.append(df2)


# Concatenate all dataframes in the list along rows
p01_F80E = pd.concat(p01_dfs1, ignore_index=True)
p01_F820 = pd.concat(p01_dfs2, ignore_index=True)

# Write the combined dataframe to a new CSV file
p01_F80E.to_csv('./dataset/P01/Presence-80029C91F80E/p01_F80E.csv', index=False)
p01_F820.to_csv('./dataset/P01/Presence-80029C91F820/p01_F820.csv', index=False)
print("p01_F80E file created successfully!")
print("p01_F820 file created successfully!")


#drop duplicates in Time column
p01_F80E = p01_F80E.drop_duplicates(subset='Time')
p01_F820 = p01_F820.drop_duplicates(subset='Time')


#Merge datasets of two sensors 

# Merge the two dataframes on the 'Time' column
p01 = pd.merge(p01_F80E, p01_F820, on='Time')
p01.drop(columns=['Unnamed: 0_y'], inplace=True)
# Rename the 'Label' columns
p01 = p01.rename(columns={'Label_x': 'label_sensor_1', 'Label_y': 'label_sensor_2'})





#fill missing data
# Convert 'Time' column to datetime
p01['Time'] = pd.to_datetime(p01['Time'], utc=True).dt.tz_convert(None)
# Generate expected time range with 5-minute frequency
expected_time_range = pd.date_range(start=p01['Time'].min(), end=p01['Time'].max(), freq='5T')

# Check for missing time rows
missing_times = expected_time_range[~expected_time_range.isin(p01['Time'])]

# If there are missing rows and they are not more than 3, fill them in
if len(missing_times) > 0 and len(missing_times) <= 3:
    for missing_time in missing_times:
        # Find the index of the row before the missing time
        idx = p01['Time'].gt(missing_time).idxmax()
        # Get values of the row before the missing time
        label_sensor_1_value = p01.at[idx, 'label_sensor_1']
        label_sensor_2_value = p01.at[idx, 'label_sensor_2']
        # Insert the missing row with values from the row before
        p01 = p01.append({'Time': missing_time,
                          'label_sensor_1': label_sensor_1_value,
                          'label_sensor_2': label_sensor_2_value}, ignore_index=True)

# Sort DataFrame by 'Time'
p01 = p01.sort_values(by='Time')

# Reset index
p01 = p01.reset_index(drop=True)


# Save the merged dataframe to a new CSV file
p01.to_csv("./dataset/P01/p01.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import os



# Load the raw dataset
def load_dataset(filename):
    return pd.read_csv(filename)


# Define functions to extract information
def extract_sleeping_episodes(df):
    episodes = []
    current_episode_start = None
    for _, row in df.iterrows():
        if row['label_sensor_1'] == 1:
            if current_episode_start is None:
                current_episode_start = row['Time']
        elif current_episode_start is not None:
            episodes.append([current_episode_start, row['Time']])
            current_episode_start = None
    if current_episode_start is not None:
        # If the last period extends until the end of the dataframe
        episodes.append([current_episode_start, df['Time'].iloc[-1]])
    return episodes

def extract_sleeping_interruptions(df, episodes):
    interruptions = []
    for i in range(len(episodes) - 1):
        end_prev_episode = episodes[i][-1]
        start_next_episode = episodes[i + 1][0]
        if start_next_episode - end_prev_episode < timedelta(hours=1):
            interruptions.append((end_prev_episode, start_next_episode))
    return interruptions

def extract_sleeping_periods(episodes, interruptions):
    periods = []
    for i, episode in enumerate(episodes):
        print(f"Episode: {i}, Length: {len(episode)}")
        period_start = episode[0]
        if interruptions and i > 0:
            if i - 1 < len(interruptions):  # Check if index is within range
                prev_interruption = interruptions[i - 1]
                period_start = prev_interruption[1]
        period_end = episode[-1]
        if interruptions and i < len(interruptions):
            next_interruption = interruptions[i]
            period_end = next_interruption[0]
        periods.append((period_start, period_end))
    return periods

def extract_outdoor_periods(df):
    periods = []
    current_period = []
    for _, row in df.iterrows():
        if row['label_sensor_1'] == 0:
            current_period.append(row['Time'])
        elif current_period:
            periods.append(current_period)
            current_period = []
    if current_period:
        periods.append(current_period)
    return periods



def calculate_activity_percentage(df):
          
    # Extract hour from 'Time' column
    df['Hour'] = df['Time'].dt.hour
    
    # Define a function to calculate activity percentage for each hour
    def calculate_hourly_activity(hour):
        # Select rows where 'Hour' equals the given hour
        hour_data = df[df['Hour'] == hour]
        
        # Count the number of rows where both sensors show 1 (sleep)
        sleep_count = ((hour_data['label_sensor_1'] == 1)).sum()
        
        # Count the number of rows where both sensors show 0 (outdoor)
        outdoor_count = ((hour_data['label_sensor_1'] == 0)).sum()
        
        # Count the number of rows where neither both sensors show 1 nor 0 (activity)
        activity_count = len(hour_data) - sleep_count - outdoor_count
        
        # Calculate percentages
        total = len(hour_data)
        sleep_percentage = (sleep_count / total) * 100
        outdoor_percentage = (outdoor_count / total) * 100
        activity_percentage = (activity_count / total) * 100
        
        return sleep_percentage, outdoor_percentage, activity_percentage
    
    # Initialize dictionaries to store percentages for each hour
    sleep_percentages = {}
    outdoor_percentages = {}
    activity_percentages = {}
    
    # Iterate over each hour of the day
    for hour in range(24):
        # Calculate percentages for the current hour
        sleep_percentage, outdoor_percentage, activity_percentage = calculate_hourly_activity(hour)
        
        # Store percentages in dictionaries
        sleep_percentages[hour] = sleep_percentage
        outdoor_percentages[hour] = outdoor_percentage
        activity_percentages[hour] = activity_percentage
        
    return sleep_percentages, outdoor_percentages, activity_percentages



sleep_percentages, outdoor_percentages, activity_percentages = calculate_activity_percentage(p01)

# Process data for each participant
def process_participant_data(filename):
    # Load dataset
    df = load_dataset(filename)
    
    # Convert date/time column to datetime format
    df['Time'] = pd.to_datetime(df['Time'], utc=True).dt.tz_convert(None)

    df['label_sensor_1'].astype(int)
    
    # Calculate activity percentages
    sleep_percentages, outdoor_percentages, activity_percentages = calculate_activity_percentage(df)

    # Apply functions to the data
    df['Date'] = df['Time'].dt.date
    dates = df['Date'].unique()

    daily_data = []
    for date in dates:
        daily_df = df[df['Date'] == date]
        sleeping_episodes = extract_sleeping_episodes(daily_df)
        interruptions = extract_sleeping_interruptions(daily_df, sleeping_episodes)
        sleeping_periods = extract_sleeping_periods(sleeping_episodes, interruptions)
        outdoor_periods = extract_outdoor_periods(daily_df)
        daily_sleep_percentages = [sleep_percentages[hour] for hour in range(24)]
        daily_outdoor_percentages = [outdoor_percentages[hour] for hour in range(24)]
        daily_activity_percentages = [activity_percentages[hour] for hour in range(24)]

        daily_data.append({
            'Date': date,
            'Sleeping_Episodes': sleeping_episodes,
            'Number_of_Sleeping_Episodes': len(sleeping_episodes),
            'Duration_of_Sleeping_Episodes': [(episode[-1] - episode[0]).total_seconds() / 3600 for episode in
                                               sleeping_episodes],
            'Sleeping_Interruptions': interruptions,
            'Number_of_Sleeping_Interruptions': len(interruptions),
            'Duration_of_Sleeping_Interruptions': [(interruption[1] - interruption[0]).total_seconds() / 3600 for
                                                    interruption in interruptions],
            'Sleeping_Periods': sleeping_periods,
            'Number_of_Sleeping_Periods': len(sleeping_periods),
            'Duration_of_Sleeping_Periods': [(period[1] - period[0]).total_seconds() / 3600 for period in
                                              sleeping_periods],
            'Outdoor_Periods': outdoor_periods,
            'Number_of_Outdoor_Periods': len(outdoor_periods),
            'Duration_of_Outdoor_Periods': [(period[-1] - period[0]).total_seconds() / 3600 for period in
                                             outdoor_periods],
            # Convert lists to dictionaries for daily percentages
            **{f'Sleep_Percentage_Hour_{hour}': daily_sleep_percentages[hour] for hour in range(24)},
            **{f'Outdoor_Percentage_Hour_{hour}': daily_outdoor_percentages[hour] for hour in range(24)},
            **{f'Activity_Percentage_Hour_{hour}': daily_activity_percentages[hour] for hour in range(24)}
        })

    # Create DataFrame for daily initial features
    daily_initial_features = pd.DataFrame(daily_data)

    return daily_initial_features



# List of filenames
filenames = ["./dataset/P02/p02.csv"]

# Process data for each participant and store in a dictionary
participant_data = {}
for filename in filenames:
    participant_id = os.path.splitext(os.path.basename(filename))[0]
    participant_data[participant_id] = process_participant_data(filename)

# Access the dataframes for each participant
daily_initial_features_p02 = participant_data['p02']

daily_initial_features_p02
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import os



# Load the raw dataset
def load_dataset(filename):
    return pd.read_csv(filename)


# Define functions to extract information
def extract_sleeping_episodes(df):
    episodes = []
    current_episode_start = None
    for _, row in df.iterrows():
        if row['label_sensor_1'] == 1 and row['label_sensor_2'] == 1:
            if current_episode_start is None:
                current_episode_start = row['Time']
        elif current_episode_start is not None:
            episodes.append([current_episode_start, row['Time']])
            current_episode_start = None
    if current_episode_start is not None:
        # If the last period extends until the end of the dataframe
        episodes.append([current_episode_start, df['Time'].iloc[-1]])
    return episodes

def extract_sleeping_interruptions(df, episodes):
    interruptions = []
    for i in range(len(episodes) - 1):
        end_prev_episode = episodes[i][-1]
        start_next_episode = episodes[i + 1][0]
        if start_next_episode - end_prev_episode < timedelta(hours=1):
            interruptions.append((end_prev_episode, start_next_episode))
    return interruptions

def extract_sleeping_periods(episodes, interruptions):
    periods = []
    for i, episode in enumerate(episodes):
        print(f"Episode: {i}, Length: {len(episode)}")
        period_start = episode[0]
        if interruptions and i > 0:
            if i - 1 < len(interruptions):  # Check if index is within range
                prev_interruption = interruptions[i - 1]
                period_start = prev_interruption[1]
        period_end = episode[-1]
        if interruptions and i < len(interruptions):
            next_interruption = interruptions[i]
            period_end = next_interruption[0]
        periods.append((period_start, period_end))
    return periods

def extract_outdoor_periods(df):
    periods = []
    current_period = []
    for _, row in df.iterrows():
        if row['label_sensor_1'] == 0 and row['label_sensor_2'] == 0:
            current_period.append(row['Time'])
        elif current_period:
            periods.append(current_period)
            current_period = []
    if current_period:
        periods.append(current_period)
    return periods



def calculate_activity_percentage(df):
          
    # Extract hour from 'Time' column
    df['Hour'] = df['Time'].dt.hour
    
    # Define a function to calculate activity percentage for each hour
    def calculate_hourly_activity(hour):
        # Select rows where 'Hour' equals the given hour
        hour_data = df[df['Hour'] == hour]
        
        # Count the number of rows where both sensors show 1 (sleep)
        sleep_count = ((hour_data['label_sensor_1'] == 1) & (hour_data['label_sensor_2'] == 1)).sum()
        
        # Count the number of rows where both sensors show 0 (outdoor)
        outdoor_count = ((hour_data['label_sensor_1'] == 0) & (hour_data['label_sensor_2'] == 0)).sum()
        
        # Count the number of rows where neither both sensors show 1 nor 0 (activity)
        activity_count = len(hour_data) - sleep_count - outdoor_count
        
        # Calculate percentages
        total = len(hour_data)
        sleep_percentage = (sleep_count / total) * 100
        outdoor_percentage = (outdoor_count / total) * 100
        activity_percentage = (activity_count / total) * 100
        
        return sleep_percentage, outdoor_percentage, activity_percentage
    
    # Initialize dictionaries to store percentages for each hour
    sleep_percentages = {}
    outdoor_percentages = {}
    activity_percentages = {}
    
    # Iterate over each hour of the day
    for hour in range(24):
        # Calculate percentages for the current hour
        sleep_percentage, outdoor_percentage, activity_percentage = calculate_hourly_activity(hour)
        
        # Store percentages in dictionaries
        sleep_percentages[hour] = sleep_percentage
        outdoor_percentages[hour] = outdoor_percentage
        activity_percentages[hour] = activity_percentage
        
    return sleep_percentages, outdoor_percentages, activity_percentages



sleep_percentages, outdoor_percentages, activity_percentages = calculate_activity_percentage(p01)

# Process data for each participant
def process_participant_data(filename):
    # Load dataset
    df = load_dataset(filename)
    
    # Convert date/time column to datetime format
    df['Time'] = pd.to_datetime(df['Time'], utc=True).dt.tz_convert(None)

    df['label_sensor_1'] = df['label_sensor_1'].astype(int)
    df['label_sensor_2'] = df['label_sensor_2'].astype(int)

    # Calculate activity percentages
    sleep_percentages, outdoor_percentages, activity_percentages = calculate_activity_percentage(df)

    # Apply functions to the data
    df['Date'] = df['Time'].dt.date
    dates = df['Date'].unique()

    daily_data = []
    for date in dates:
        daily_df = df[df['Date'] == date]
        sleeping_episodes = extract_sleeping_episodes(daily_df)
        interruptions = extract_sleeping_interruptions(daily_df, sleeping_episodes)
        sleeping_periods = extract_sleeping_periods(sleeping_episodes, interruptions)
        outdoor_periods = extract_outdoor_periods(daily_df)
        daily_sleep_percentages = [sleep_percentages[hour] for hour in range(24)]
        daily_outdoor_percentages = [outdoor_percentages[hour] for hour in range(24)]
        daily_activity_percentages = [activity_percentages[hour] for hour in range(24)]

        daily_data.append({
            'Date': date,
            'Sleeping_Episodes': sleeping_episodes,
            'Number_of_Sleeping_Episodes': len(sleeping_episodes),
            'Duration_of_Sleeping_Episodes': [(episode[-1] - episode[0]).total_seconds() / 3600 for episode in
                                               sleeping_episodes],
            'Sleeping_Interruptions': interruptions,
            'Number_of_Sleeping_Interruptions': len(interruptions),
            'Duration_of_Sleeping_Interruptions': [(interruption[1] - interruption[0]).total_seconds() / 3600 for
                                                    interruption in interruptions],
            'Sleeping_Periods': sleeping_periods,
            'Number_of_Sleeping_Periods': len(sleeping_periods),
            'Duration_of_Sleeping_Periods': [(period[1] - period[0]).total_seconds() / 3600 for period in
                                              sleeping_periods],
            'Outdoor_Periods': outdoor_periods,
            'Number_of_Outdoor_Periods': len(outdoor_periods),
            'Duration_of_Outdoor_Periods': [(period[-1] - period[0]).total_seconds() / 3600 for period in
                                             outdoor_periods],
            # Convert lists to dictionaries for daily percentages
            **{f'Sleep_Percentage_Hour_{hour}': daily_sleep_percentages[hour] for hour in range(24)},
            **{f'Outdoor_Percentage_Hour_{hour}': daily_outdoor_percentages[hour] for hour in range(24)},
            **{f'Activity_Percentage_Hour_{hour}': daily_activity_percentages[hour] for hour in range(24)}
        })

    # Create DataFrame for daily initial features
    daily_initial_features = pd.DataFrame(daily_data)

    return daily_initial_features



# List of filenames
filenames = ["./dataset/P01/p01.csv", "./dataset/P03/p03.csv", "./dataset/P05/p05.csv"]

# Process data for each participant and store in a dictionary
participant_data = {}
for filename in filenames:
    participant_id = os.path.splitext(os.path.basename(filename))[0]
    participant_data[participant_id] = process_participant_data(filename)

# Access the dataframes for each participant
daily_initial_features_p01 = participant_data['p01']
daily_initial_features_p03 = participant_data['p03']
daily_initial_features_p05 = participant_data['p05']
```

```python
def sleep_related_features(df):
    total_sleep_durations = []
    interruptions = []
    interruption_durations = []
    sleep_episode_durations = []
    longest_period_durations = []
    longest_episode_durations = []

    for _, row in df.iterrows():
        total_sleep_durations.extend(row['Duration_of_Sleeping_Periods'])
        interruptions.append(row['Number_of_Sleeping_Interruptions'])  # Append the integer value
        interruption_durations.extend(row['Duration_of_Sleeping_Interruptions'])
        sleep_episode_durations.extend(row['Duration_of_Sleeping_Episodes'])
        if row['Duration_of_Sleeping_Periods']:  # Check if the list is not empty
            longest_period_durations.append(max(row['Duration_of_Sleeping_Periods']))
        if row['Duration_of_Sleeping_Episodes']:  # Check if the list is not empty
            longest_episode_durations.append(max(row['Duration_of_Sleeping_Episodes']))

    avg_sleep_duration = np.mean(total_sleep_durations)
    std_sleep_duration = np.std(total_sleep_durations)

    avg_interruptions = np.mean(interruptions)
    std_avg_interruptions = np.std(interruptions)

    avg_interruption_duration = np.mean(interruption_durations)
    std_avg_interruption_duration = np.std(interruption_durations)

    avg_sleep_episode_duration = np.mean(sleep_episode_durations)
    std_sleep_episode_duration = np.std(sleep_episode_durations)

    avg_longest_period_duration = np.mean(longest_period_durations)
    avg_longest_episode_duration = np.mean(longest_episode_durations)

    return {
        'Avg_Total_Sleep_Duration': avg_sleep_duration,
        'Std_Total_Sleep_Duration': std_sleep_duration,
        'Avg_Num_Interruptions': avg_interruptions,
        'Std_Avg_Num_Interruptions': std_avg_interruptions,
        'Avg_Interruption_Duration': avg_interruption_duration,
        'Std_Avg_Interruption_Duration': std_avg_interruption_duration,
        'Avg_Sleep_Episode_Duration': avg_sleep_episode_duration,
        'Std_Sleep_Episode_Duration': std_sleep_episode_duration,
        'Avg_Longest_Period_Duration': avg_longest_period_duration,
        'Avg_Longest_Episode_Duration': avg_longest_episode_duration
    }

# Calculate sleep-related features for each participant
sleep_features_p01 = sleep_related_features(daily_initial_features_p01)
sleep_features_p02 = sleep_related_features(daily_initial_features_p02)
sleep_features_p03 = sleep_related_features(daily_initial_features_p03)
sleep_features_p05 = sleep_related_features(daily_initial_features_p05)

# Convert the dictionaries to DataFrames
sleep_features_df_p01 = pd.DataFrame([sleep_features_p01])
sleep_features_df_p02 = pd.DataFrame([sleep_features_p02])
sleep_features_df_p03 = pd.DataFrame([sleep_features_p03])
sleep_features_df_p05 = pd.DataFrame([sleep_features_p05])




sleep_features_df_p01
sleep_features_df_p02
sleep_features_df_p03
sleep_features_df_p05
```

Feature engineering of activity related features (P01, P03, P05)

```python
import numpy as np
import pandas as pd

def activity_related_features(df):
    total_home_durations = []
    outdoor_periods = []

    for _, row in df.iterrows():
        total_home_durations.extend(row['Duration_of_Sleeping_Periods'])
        outdoor_periods.append(row['Number_of_Outdoor_Periods'])  # Append the integer value

    avg_home_duration = 24 - np.mean(total_home_durations)
    std_home_duration = 24 - np.std(total_home_durations)

    avg_outdoor_duration = np.mean(total_home_durations)
    std_outdoor_duration = np.std(total_home_durations)

    avg_outdoor_periods = np.mean(outdoor_periods)
    std_avg_outdoor_periods = np.std(outdoor_periods)

    ratio_home_outdoor = avg_home_duration / avg_outdoor_duration

    return {
        'Avg_Home_Duration': avg_home_duration,
        'Std_Home_Duration': std_home_duration,
        'Avg_Outdoor_Duration': avg_outdoor_duration,
        'Std_Outdoor_Duration': std_outdoor_duration,
        'Avg_Outdoor_Periods': avg_outdoor_periods,
        'Std_Avg_Outdoor_Periods': std_avg_outdoor_periods,
        'Ratio_Home_Outdoor': ratio_home_outdoor,
    }

# Calculate activity-related features for each participant
activity_features_p01 = activity_related_features(daily_initial_features_p01)
activity_features_p02 = activity_related_features(daily_initial_features_p02)
activity_features_p03 = activity_related_features(daily_initial_features_p03)
activity_features_p05 = activity_related_features(daily_initial_features_p05)

# Convert the dictionaries to DataFrames
activity_features_df_p01 = pd.DataFrame([activity_features_p01])
activity_features_df_p02 = pd.DataFrame([activity_features_p02])
activity_features_df_p03 = pd.DataFrame([activity_features_p03])
activity_features_df_p05 = pd.DataFrame([activity_features_p05])

# Display the resulting DataFrames
activity_features_df_p01
activity_features_df_p02
activity_features_df_p03
activity_features_df_p05
```
